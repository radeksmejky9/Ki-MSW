```python
import random
import matplotlib.pyplot as plt

width_large = 1
height_large = 1
width_small = 1
height_small = 0.5

num_points = 10000

inside_rectangle = 0
points_inside_x = []
points_inside_y = []
points_outside_x = []
points_outside_y = []

for _ in range(num_points):
    x = random.uniform(0, width_large)
    y = random.uniform(0, height_large)

    if 0 <= x <= width_small and 0 <= y <= height_small:
        points_inside_x.append(x)
        points_inside_y.append(y)
    else:
        points_outside_x.append(x)
        points_outside_y.append(y)

rectangle_area_estimate = (len(points_inside_x) / num_points) * (width_large * height_large)
```

```python
# Vykreslení bodů uvnitř a vně obdélníka
plt.figure(figsize=(6, 6))
plt.scatter(points_inside_x, points_inside_y, color="blue", s=1, label="Body uvnitř obdélníka")
plt.scatter(points_outside_x, points_outside_y, color="red", s=1, label="Body vně obdélníka")
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel("X")
plt.ylabel("Y")
plt.title(f"Odhad plochy obdélníka: {rectangle_area_estimate}")
plt.legend(loc="upper right").get_frame().set_alpha(None)
plt.show()
```
